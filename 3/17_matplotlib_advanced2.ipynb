## 3. Text, Fonts and LaTeX
 `Matplotlib` allows complete control over text properties (font, size, position, color...). With `Matplotlib` it is possible to display mathematical symbols in any figure using **LaTeX** commands. <br>
 Some text functions were already introduced earlier, but here is a complete list of `pyplot` functions that can be used to insert text. We present them with the objects '**`fig`**' and '**`ax`**'. <br>
* `text` <=> `ax.text`: Inserts text at any position in '`ax`'. <br>
* `xlabel` <=> `ax.set_xlabel`: Adds a label to the x-axis of '`ax`'. <br>
* `ylabel` <=> `ax.set_ylabel`: Adds a label to the y-axis of '`ax`'. <br>
* `title` <=> `ax.set_title`: Adds a title to '`ax`'. <br>
* `figtext` <=> `fig.text`: Adds a title to '`fig`'. <br>
* `suptitle` <=> `fig.suptitle`: Adds a centered title to '`fig`'. <br>
* `annotate` <=> `ax.annotate`: Adds an annotation and arrow to '`ax`'. <br>
 All these commands return a `Text` object. The text can be customized by passing arguments to functions or by using `matplotlib.artist.setp` or `plt.setp`. <br>
 <span style="color:#09b038; text-decoration : underline"> Example: Text customization </span><br>
```python
t = plt.xlabel('A label', fontsize=16, color='green')
```
 is equivalent to <br>
```python
t = plt.xlabel('A label') 
setp(t, fontsize=16, color='green')
```
 The text font can be changed with the parameter `family`. <br>
 There are 6 available font families for the texts: 'serif', 'sans-serif', 'cursive', 'fantasy' and 'monospace'. <br>
 The font style can also be changed with `fontstyle` and can be: 'normal', 'italic' or 'oblique'. <br>
 One can access the text objects of a graph with functions like `ax.get_xticklabels`, which returns the labels of the x-axis markers. <br>
 With object parameters it is possible to assign the same properties to all objects in the same group. <br>
 <span style="color:#09b038; text-decoration : underline"> Example: Group objects </span><br>
```python
for t in ax.get_xticklabels():
    t.set_fontsize(5.)
```
 With `setp`: <br>
```python
plt.setp(ax.get_xticklabels(), fontsize=5.)
```
 The font size of all x-axis labels was changed. <br>

## 4. Contour Plots
 The function `plt.contour` allows the display of contour plots. <br>
 In the case of a function with two variables, we also call them contour lines or isolines. On such a line, the function value is constant. <br><br>
 Mathematically speaking, a contour line represents f(x, y) = L for various constant values of L. <br>
 To draw contour lines, two number sequences **x** and **y** with sizes **m** and **n** respectively are instantiated, representing coordinate vectors. Then a two-dimensional grid must be created that stores the coordinate matrices from the combination of **x** and **y**. This is done with the function **`np.meshgrid`**, which returns two arrays **X** and **Y** of size **n x m**. <br><br>
 Then standard arithmetic operations can be performed on these matrices to create a surface **Z** that contains the values of the function to be drawn. <br>
 The command `plt.contour(X,Y,Z)` draws contour lines using X and Y as coordinates. It is possible to specify the number of lines to be drawn or even the values of the lines. <br>
 <span style="color:#09b038; text-decoration : underline"> Example: Contour curves </span><br>
```python
x = np.arange(-1,1,0.05)
y = np.arange(-2, 2, 0.05)
X,Y = np.meshgrid(x,y)
Z = Y**2 - X**2
plt.contour(X,Y,Z)
```

 It is also possible to display the value of the lines in the graph using **`plt.clabel`**. <br>

## 5. Polar Plots
 Polar plots use a completely different coordinate system. Up to now we have only used Cartesian coordinates, where two orthogonal axes intersect at the origin and define precise directions for positive and negative values. A polar system is a 2D coordinate system where the position of a point is expressed by radius and angle. <br><br>
 This system is used when the relationship between two points can be better expressed in this coordinate system. <br><br>
 A point is represented by a radial and an angular component. The radial coordinate, often referred to as **r**, is called radius. It expresses the distance to the origin (the pole). The angular coordinate, denoted as θ, is called polar angle. It expresses the measure of the angle between the point and the 0° half-line (the polar axis, similar to the x-axis) in trigonometric direction (positive direction). <br><br>
 The function `polar` from `Matplotlib` displays polar plots. Its parameters are 2 lists of equal size, `theta` for the angular coordinates and `r` for the radial coordinates. It is the equivalent to `plot` for polar plots. <br><br>
 <span style="color:#09b038; text-decoration : underline"> Example: Polar coordinates </span><br>
```python
theta = np.arange(0, 2, 1/180)*np.pi 
plt.polar(3*theta, theta/5)
plt.polar(theta, np.cos(4*theta))
plt.polar(theta, [1.4]*len(theta)); 
```

 In the example we define **`theta`** as a sequence of 360 equidistant values between 0 and 2$\pi$. The second line draws a spiral. Below that we draw a **rosette**. <br>
 In the polar coordinate system, r must remain constant (here equal to 1.4) to draw a circle. <br>
