####
To add an annotation with a descriptive arrow pointing to a specific point in the graph, use `plt.annotate`.
This method takes the following arguments: <br>
- the text you want to display
- `xy`, which specifies the coordinates where the point to be annotated is located
- `xytext`, which specifies the coordinates of the point where the text begins
- `arrowprops`, which specifies the properties of the annotation arrow between { }: colors, arrow size, arrow style, etc.

<span style="color:#09b038; text-decoration : underline">Example: Using the **plt.annotate** command</span><br>
```python
plt.annotate(‘Boundary’, xy=(1, 2), xytext=(1, 2.5), arrowprops={‘facecolor’:'blue'})
```
Displays a blue arrow pointing to the coordinate point (1, 2) and displays the text ‘Boundary’ at the point (1, 2.5).

````python
# Deine Lösung:



````

#### Lösung:

````python
plt.grid(True)

x = np.arange(0., 5., .01)

y = np.sin(2*np.pi*x) * np.exp(-x)
plt.plot(x, y)
plt.plot(x, np.exp(-x))
plt.plot(x, -np.exp(-x))

plt.annotate('Asymptote: exp(-x)', xy=(1.5, 0.25), xytext=(2, 0.5), 
            arrowprops={'facecolor':'green'})
# An arrow is created pointing to the point (1.5, 0.25) with the annotation “Asymptote: exp(-x)”
# starting at the point (2, 0.5). Then we give it the color green.

plt.annotate('Asymptote: -exp(-x)', xy=(1.5, -0.25), xytext=(2, -0.5), 
            arrowprops={'facecolor':'red'})
````

````python
x = np.linspace(-6, 2, 40)
# A sequence of 40 points is created in the interval [-6, 2].

plt.grid(True, linestyle='--')
# A grid is added to the graph.

plt.plot(x, x**2, 'b', label='$x^2$')
# We draw the line graph x^2 and give it the corresponding label for the legend.

plt.plot(x, np.exp(-x), 'g--', label='$exp(-x)$')
# The line graph exp(-x) is drawn as a dashed line and given the corresponding label for the legend.

plt.axis([-5,2,0,25])
# We alter the amplitude of the axes.

plt.plot([-3, -2, -2, -3, -3], [5, 5, 10, 10, 5], 'r', alpha=0.6)
# We create the square that surrounds the part of the graphic that we will reproduce
# The argument ‘alpha’ specifies the percentage of opacity of the drawing (1 opaque, 0 invisible transparent)

plt.annotate('Zoom', xy=(-1.8, 7.5), xytext=(-0.5, 7.5), 
            arrowprops={'facecolor':'red'})
# The red arrow is created with the text ‘Zoom’ pointing to the point (-1.8, 7.5).

plt.title('My summary chart')
# We add a title

plt.legend()
# show legend

plt.axes([.55, 0.4, .2, .2])
# A new graph is created within the previous one,
# whose lower left corner starts at point (0.55, 0.4) in relative distance,
# where 0 represents the origin and 1 represents the end of the axis.
# This plot will have a width and height of 20% of the width
# and height of the original plot.

# From now on, the ‘pyplot’ commands only apply to this new plot.

plt.plot(x, x**2, 'b')
plt.plot(x, np.exp(-x), 'g--')
# We draw the same line charts as before.

plt.axis([-3,-2,5,10])
# But we limit the axes to the coordinates of the square we have drawn.

plt.xticks([])
plt.yticks([])
# Delete the graduations on the new graphic
````
