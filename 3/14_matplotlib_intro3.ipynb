## 1. Boxplots

Boxplots are highly valued and frequently used diagrams, especially for descriptive analysis of continuous data. <br>

The **`plt.boxplot`** function allows you to display boxplots in a diagram. <br>

It takes one or more sequences of values summarized in a list as an argument. <br>

The box shows the boundaries of the first and last quartiles, and the red bar in the middle shows the median. The whiskers extend to the most extreme value within the boundary of 1.5 times the box height. <br>

The points beyond the whiskers are represented by +; they are *outliers*. <br>

<span style="color:#09b038; text-decoration : underline">Examples:</span>
```python
plt.boxplot([[1, 2, 3, 4, 5],[7, 5, 8, 4, 9, 5, 7]])
```
creates the following graph: <br>

## <i class="fas fa-chart-pie"></i> 2. Pie charts

Pie charts are charts divided into sectors (*wedges*). This is an effective way to present information when you are primarily interested in comparing one sector to the entire pie chart rather than to another sector. <br>

The **`pie`** function allows you to create pie charts from a sequence **`X`**. <br>
Each element **$x_{i}$** of **`X`** is represented by a proportional share of **$x_{i}$/sum(X)**. <br>

If the sum of the elements of **`X`** is less than 1, the pie chart is displayed directly with the values of **`X`**, no normalization is performed, and an empty space remains. <br>

<div class="alert alert-success">
<i class="fa fa-question-circle"></i> &emsp; 
Pie charts are best displayed when the figure is enlarged and <b>square</b>. <br>
When using a <b>pyplot</b> function, a figure in standard format is created automatically. <br>
The figure can be created in the dimensions <b>(h,l)</b> using the command <b>plt.figure(figsize=(h,l))</b>.
</div>

<div class="alert alert-info">
<i class="fa fa-info-circle"></i> &emsp;
Add the argument <b>labels</b> to <b>plt.pie</b> to give names to the different parts of the pie chart.
</div>

## 3. Subplots

It can be useful to visualize multiple graphs in a single figure, especially to compare results or show how different variables relate to each other. <br>

<img src="../imgs/subplot_111.png" style="height:350px">


Multiple graphs can be displayed in a grid in the same Matplotlib figure thanks to the **`subplot`** function, which automatically creates a figure and divides it as desired. <br>

The **`subplot`** function takes as arguments: the number of rows in the figure (*numrows*), the number of columns (*numcols*), and the number of the graph you want to position yourself on (between 1 and *numrows* x *numcols*). <br>

Commas are optional if $numrows \times numcols < 10$. Thus, `subplot(211)` is identical to `subplot(2, 1, 1)`. <br>

<span style="color:#09b038; text-decoration : underline">Example:</span> The command `plt.subplot(4,3,2)` at the beginning of the cell creates and divides a figure into 4 x 3 plot spaces (4 rows and 3 columns) and selects the 2nd space for the following plot instructions. To display a graph in the next space, simply write `plt.subplot(4,3,3)` followed by the desired **plt** functions. <br>

<div class="alert alert-success">
<i class="fa fa-question-circle"></i> &emsp; 
The numbers of the plots are counted row by row. For example, the plot in the second row and third column is plot no. 6.
</div>

<div class="alert alert-success">
<i class="fa fa-question-circle"></i> &emsp; 
When using a <b>pyplot</b> function, a figure in the standard format is automatically created. <br>
It is possible to create a figure with dimensions <b>(h,l)</b> using the command <b>plt.figure(figsize=(h,l))</b>.
</div>

#### 
The graphical methods of Pandas DataFrames do not directly allow you to display multiple different graphs within the same figure. <br>
However, it is useful to know that some methods, such as `plot`, include the `subplots` parameter, which, when set to `True`, divides the figure into as many graphs as there are variables. The `layout` parameter allows you to choose the layout of the plots. <br>

Here is an example of creating graphs in the same figure with Pandas: <br>

```python
df.plot(y=[‘Product1’, ‘Product2’, ‘Returns’, ‘Sales’], subplots=True, layout=(2,2),
        style=[‘b--’, ‘m:p’, ‘g-.’, ‘c-d’], figsize=(7,7))
```

With **`Matplotlib`**, it is also possible to insert a graph within another graph, thanks to the **`axes`** function. <br>

All you have to do is pass it a list with the `x` and `y` positions of the bottom left corner of the graph to be inserted (values between `0` and `1`, regardless of the selected axes), as well as its dimensions (width and height, between `0` and `1`). <br>

For greater visibility, it is recommended to give this axis a different background color by adding the argument `facecolor`.
For example, `plt.axes([.5, 0.6, .2, .2], facecolor=‘gray’)` creates a graph with a gray background, whose lower left point on the x-axis is in the middle of the current graph, on the y-axis at 60% of its height, and whose width and height are 20% of the current graph.

<div class="alert alert-success">
<i class="fa fa-question-circle"></i> &emsp; 
For aesthetic reasons, it is preferable to remove the graduations from the axes of graphs inserted into others. 
The commands <b>plt.xticks([ ])</b> and <b>plt.yticks([ ])</b> are sufficient to delete the x- and y-axes, respectively.
</div>
