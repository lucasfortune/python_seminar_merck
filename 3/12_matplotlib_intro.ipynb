## 1. Creating Line Charts

A line chart is a set of points with coordinates (x, y), between which a line is drawn: The more points used, the smoother the plot looks. <br>

The **`plot`** function allows you to draw lines that connect points, whose X and Y coordinates are provided in lists or arrays. To draw a graph with values '`x`' on the X-axis and values '`y`' on the Y-axis, we write: 

```python
x = [1,2,3]
y = [1,2,3]
plt.plot(x,y)
```

#### 
Matplotlib also offers the possibility to change the style of elements in charts, which can be very useful to further distinguish them and even add additional meaning (such as a specific color). <br>

For each pair (x,y) there is a third optional argument to set the style and color of the line. E.g. for a red "r" solid "-" line:<br>

><code>plt.plot(<span style="color:red">x,y</span>,<span>"-r"</span>)</code>

````python
# Your solution:



````

You can specify the style of a line chart in **one string** in this order: the color, the line style and the markers. <br>
For example: <br>
* `"r-*"` for a red solid line with star markers <br>
* `"y:d"` for a yellow dotted line with diamond markers, etc. <br>

When drawing a line, it is also possible to adjust its width, thanks to the *`linewidth`* parameter of the `plot` function. <br>
So the command: `plt.plot(x, y, "r--", linewidth=5, marker='o')` displays a red dashed line with width 5 and circle markers. <br>

The position of the legend can be chosen with the `loc` argument, which can be `'best'`, `'lower right'`, `'upper center'`, `'center left'`, etc. <br>

## Summary:

- To create line charts with *Matplotlib*, we use the ```plt.plot()``` function, which works with ```lists```, ```numpy arrays```, and ```Pandas Series```. <br>
- A single plot command can draw multiple lines by passing multiple x,y pairs: ```plt.plot(x1,y1, x2,y2, x3,y3)```. <br>
- Matplotlib offers diverse formatting options for your plots: <br>

    - Line styles ```'-', '--', ':', '-.'``` for different line types
    - Colors ```'r', 'g', 'b', etc.``` for different lines
    - Markers ```'o', '*', '+', etc.``` for data points
    - Line width with ```linewidth``` for emphasizing certain lines <br>


- For better readability of your charts, you can add various elements: <br>

    - Legends with ```plt.legend()``` and the ```label``` parameter
    - Axis labels with ```xlabel``` and ```ylabel```
    - Titles with ```title```
    - Grid with ```plt.grid(True)```
    - Axis limits with ```xlim``` and ```ylim``` <br>


- *Pandas* DataFrames offer simplified syntax for plotting with Matplotlib in the background: <br>

    - Direct plotting of multiple columns with ```df.plot(x='column1', y=['column2', 'column3'])```
    - Easy style customization with the ```style``` parameter
    - Automatic legend creation based on column names <br>
