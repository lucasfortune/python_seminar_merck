## 1. Images
 In mathematics, an image is a function that quantifies the light intensity of each point of an image. <br>
 In a black and white image, the intensity is the grayscale value: the darker a point is, the lower its grayscale value. <br>
 
 In a color image, a point is quantified by its intensity of red, green and blue. <br>
 A digital image is sampled: it is divided according to a regular grid and each image cell is assigned an intensity, which is called a pixel. <br>
 
 The grayscale values or color values in the RGB model are integers between 0 and 255. <br>
 The command `plt.imread` allows converting an image into a three-dimensional array (or 2D if the image is in black and white). <br>
 
 For an RGB image, it is a 3D array of 3 tables of pixels in the range from 0 to 255. <br>
 
 The command `plt.imshow` allows displaying an image array within a graph. <br>

### 2.4 Wireframe Models
 A *wireframe model* is a three-dimensional shape that consists only of "wires" (lines and vertices) and has no surfaces or textures. <br>
 
 The *wireframe models* take a grid of values and project it into three-dimensional space for easy visualization. <br>
 
 The method **`plot_wireframe`** allows the display of *wireframe models* and takes as arguments three 2-dimensional coordinate grids. <br>
 
#### 2.4.1 Exercises:
> (a) Create the vector **u** with a sequence of 30 numbers between 0 and $\pi$ <br>
>
> (b) Create the vector **v** with a sequence of 30 numbers between 0 and $2\pi$ <br>
>
> (c) Create the table **x** with the matrix product $sin(u) (sin(v)^T)$ using the *numpy* function **`outer`** <br>
>
> (d) Create the table **y** with the matrix product $sin(u) (cos(v)^T)$ using the *numpy* function **`outer`** <br>
>
> (e) Create the table **z** with the matrix product $cos(u) (\mathbb{1}_{30})^T$, where $\mathbb{1}_{30}$ represents a vector of size 30 filled with 1 <br>
>
> (f) Display the *wireframe model* from **x**, **y**, **z**

````python
# Your Solution:



````

#### Solution:

````python
u = np.linspace(0, np.pi, 30)
v = np.linspace(0, 2 * np.pi, 30)

x = np.outer(np.sin(u), np.sin(v))
y = np.outer(np.sin(u), np.cos(v))
z = np.outer(np.cos(u), np.ones(30))

fig = plt.figure(figsize=(8,6))
ax = plt.axes(projection='3d')

ax.plot_wireframe(x, y, z);
````

### 2.5 Surface Plots
 Surface plots are sometimes useful to visualize complex functions and their contours in 3D. <br>
 
 The method **`plot_surface`** allows the display of a 3D representation of the surface of a function, which is colored according to its values. <br>
 
#### 2.5.1 Exercises:
> (a) Create the matrix **x**, defined by the following matrix product: `np.outer(np.linspace(-2, 2, 30), np.ones(30))` <br>
>
> (b) Create **y**, the transpose of **x** <br>
>
> (c) Create **z** as the function: $ z = \cos(x^2 + y^2)$ <br>
>
> (d) Display a *surface plot* of **x**, **y** and **z** and add the argument `cmap=plt.cm.Spectral_r` to adjust the colors used

````python
# Your Solution:



````

#### Solution:

````python
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.T
z = np.cos(x ** 2 + y ** 2)

fig = plt.figure(figsize=(8,6))
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z, cmap=plt.cm.Spectral_r);
````
