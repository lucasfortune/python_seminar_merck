# What is Matplotlib?

Matplotlib is the most comprehensive plotting library for Python. It was originally developed as a MATLAB-like plotting interface and has become the de facto standard for visualizations in Python. <br>

Important features of Matplotlib are: <br>
- Generation of publication-ready graphics <br>
- Large selection of plot types and customization options <br>
- Good integration with NumPy and Pandas <br>
- Export to various file formats (PNG, PDF, SVG, etc.) <br>
- Support for interactive plots <br>

Matplotlib is frequently used in scientific applications because it allows precise control over every aspect of a visualization. <br>

## 1 Basic Plot Creation

Matplotlib offers various basic plot types: <br>

```python
import matplotlib.pyplot as plt
import numpy as np

# Daten generation
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Simple Line Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine')
plt.title('Simple line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.legend()
plt.show()
```

## 2 Various Plot Types

Matplotlib supports various plot types: <br>

```python
# Data for various plots
x = np.linspace(0, 10, 30)
y1 = np.sin(x)
y2 = np.random.randn(30)
y3 = np.random.rand(30)

# Create multiple plots
plt.figure(figsize=(15, 10))

# Scatter diagram
plt.subplot(2, 2, 1)
plt.scatter(x, y2, c='red', alpha=0.5, label='Scatter')
plt.title('Scatter')
plt.legend()

# Bar diagram
plt.subplot(2, 2, 2)
plt.bar(x[:10], y3[:10], label='Bars')
plt.title('Bars')
plt.legend()

# Histogram
plt.subplot(2, 2, 3)
plt.hist(y2, bins=15, alpha=0.7, label='Histogram')
plt.title('Histogram')
plt.legend()

# Step Plot
plt.subplot(2, 2, 4)
plt.step(x, y1, label='Steps')
plt.title('Step Plot')
plt.legend()

plt.tight_layout()
plt.show()
```

## 3 Adapting Plots

Detailed control over the look of the plots: <br>

```python
# Plot with multiple adaptations
plt.figure(figsize=(12, 6))

# Different line types and markers
plt.plot(x, np.sin(x), 'r--', label='Sine', linewidth=2)
plt.plot(x, np.cos(x), 'b.-', label='Cosine', markersize=10)
plt.plot(x, -np.sin(x), 'g:', label='-Sine', linewidth=3)

# Axis labeling with LaTeX
plt.xlabel(r'$\theta$ (rad)', fontsize=12)
plt.ylabel(r'$f(\theta)$', fontsize=12)

# Title with different font
plt.title('Trigonometric Functions', 
         fontsize=14, 
         fontweight='bold')

# Grid
plt.grid(True, linestyle='--', alpha=0.7)

# Legend
plt.legend(loc='upper right', 
          framealpha=0.9, 
          shadow=True)

plt.show()
```

## 4 Statistical Visualizations

Plots for statistical data: <br>

```python
# Daten generation
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(2, 1.5, 1000)

plt.figure(figsize=(15, 5))

# Box Plot
plt.subplot(131)
plt.boxplot([data1, data2], labels=['Set 1', 'Set 2'])
plt.title('Box Plot')

# Violin Plot
plt.subplot(132)
plt.violinplot([data1, data2])
plt.title('Violin Plot')

# KDE Plot
plt.subplot(133)
from scipy import stats
x_range = np.linspace(-5, 8, 200)
kde1 = stats.gaussian_kde(data1)
kde2 = stats.gaussian_kde(data2)
plt.plot(x_range, kde1(x_range), label='Set 1')
plt.plot(x_range, kde2(x_range), label='Set 2')
plt.title('KDE Plot')
plt.legend()

plt.tight_layout()
plt.show()
```

## 5 3D-Plots

Matplotlib can also create 3D visualizations: <br>

```python
from mpl_toolkits.mplot3d import Axes3D

# 3D Daten generation
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# 3D Surface Plot
fig = plt.figure(figsize=(12, 5))

ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_title('Surface Plot')
fig.colorbar(surf)

# 3D Wireframe
ax2 = fig.add_subplot(122, projection='3d')
wire = ax2.plot_wireframe(X, Y, Z, rstride=5, cstride=5)
ax2.set_title('Wireframe Plot')

plt.tight_layout()
plt.show()
```

## 6 Specialized Plots

Matplotlib also offers specialized plots: <br>

```python
# Polar Plot
plt.figure(figsize=(12, 5))

plt.subplot(121, projection='polar')
theta = np.linspace(0, 2*np.pi, 100)
r = 2*np.cos(4*theta)
plt.plot(theta, r)
plt.title('Polar Plot')

# Contour plot
plt.subplot(122)
plt.contour(X, Y, Z, levels=15, cmap='RdGy')
plt.colorbar(label='z-Wert')
plt.title('Contour plot')

plt.tight_layout()
plt.show()
```

## 7 Style and Design

Matplotlib offers various themes and designs: <br>

```python
# Show available styles
print("Available styles:", plt.style.available)

# Demonstrate different styles
data = np.random.randn(1000)

styles = ['default', 'ggplot', 'bmh']
plt.figure(figsize=(15, 10))

for i, style in enumerate(styles, 1):
    with plt.style.context(style):
        plt.subplot(2, 2, i)
        plt.hist(data, bins=30)
        plt.title(f'Style: {style}')

plt.tight_layout()
plt.show()
```
